2. Cree un array cuyos elementos cuyos elementos sean los enteros pares en [1,100] y en orden decreciente. Muestre los 10 últimos por pantalla.

3) Dado el array
2 4 5 6
0 3 7 4
8 8 5 2
1 5 6 1
Seleccione con una instrucción el subarray de elementos
0 3 7 4
Después, seleccione el subarray de elementos
2 5
8 5

```python
import numpy as np

# Creamos el array
array = np.array([[2, 4, 5, 6],
                  [0, 3, 7, 4],
                  [8, 8, 5, 2],
                  [1, 5, 6, 1]])

# Seleccionamos el subarray 0 3 7 4
subarray1 = array[1]

# Seleccionamos el subarray 2 5 y 8 5
subarray2 = array[[0,2], [1,2]]

# Imprimimos los resultados
print("Subarray 1:", subarray1)
print("Subarray 2:\n", subarray2)
```

4) Dados los arrays a = [1, 4, 2, 7] y b = [1, 3, 2, 9], obtenga la media aritmética de la diferencia (a-b).

```python
import numpy as np

# Definimos los arrays a y b
a = np.array([1, 4, 2, 7])
b = np.array([1, 3, 2, 9])

# Calculamos la diferencia entre a y b
diferencia = a - b

# Calculamos la media de la diferencia
media = np.mean(diferencia)

# Imprimimos el resultado
print("La media aritmética de la diferencia (a-b) es:", media)
```

5) Tengo valores de cordenadas (x, y) en las columnas del array
1.33 4.5
30.0 10.7
70.2 0.5

```python
import numpy as np

# Definimos los valores de coordenadas en las columnas del array
columnas = np.array([[1.33, 4.5],
                     [30.0, 10.7],
                     [70.2, 0.5]])

# Obtenemos los valores de x e y como arrays separados
x = columnas[:, 0]
y = columnas[:, 1]

# Imprimimos los resultados
print("Valores de coordenadas en las columnas del array:\n", columnas)
print("Valores de x:", x)
print("Valores de y:", y)
```

Agregue a este array las coordenadas (37.1, -3.6). Muestre en pantalla las dimensiones del nuevo array.

```python
import numpy as np

# Definimos el array original
columnas = np.array([[1.33, 4.5],
                     [30.0, 10.7],
                     [70.2, 0.5]])

# Definimos las coordenadas que queremos agregar
nueva_coordenada = np.array([37.1, -3.6])

# Agregamos las nuevas coordenadas al array original
columnas = np.append(columnas, [nueva_coordenada], axis=0)

# Imprimimos las dimensiones del nuevo array
print("El nuevo array tiene dimensiones:", columnas.shape)
```
